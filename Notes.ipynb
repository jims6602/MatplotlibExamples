Setting Update sample data to work with

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

# Function

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Linspace Zero to Five divided into eleven parts')
plt.show()
```

```python
plt.subplot(1,3,1) # layout is one row and two columns. This is first plot
plt.plot(x,y,'r') # color is red

plt.subplot(1,3,2) # layout is one row and two columns. This is second plot
plt.plot(y,x,'b') # color is blue

plt.subplot(1,3,3) # layout is one row and two columns. This is third plot
plt.plot(y,x,'g') # color is green
```

# Object Oriented 

```python
fig = plt.figure()

# add_axes parameter List [x Starting point, y starting point, width, hieght]
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('LARGE PLOT')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.plot(x,y,'r')# r for red

axes2 = fig.add_axes([.2, .5, .4 , .3])
axes2.set_title('SMALLER PLOT')
axes2.plot(y,x,'b')# b for blue
```

8.43 Matplotlib Part 2 (Pierian Data)

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI

```python
#dpi = dots per inch

fig = plt.figure(figsize=(8,2), dpi=100) # eight by two size of the graph

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2)) 

axes[0].plot(x,y)


axes[1].plot(y,x)
plt.tight_layout()
```
